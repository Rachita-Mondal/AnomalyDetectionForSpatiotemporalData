```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
LRSTSS = pd.read_csv("LRSTSS_varying_dur.csv")
LRSTS = pd.read_csv("LRSTS_varying_dur.csv")
HORPCA = pd.read_csv("HORPCA_varying_dur.csv")
LRSSS =  pd.read_csv("LRSSS_varying_dur.csv")
```

```python
   
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1) 
    sns.lineplot(data=LRSTSS[["seed", "duration", "roc_auc"]], x="duration", y="roc_auc",marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSTS[["seed", "duration", "roc_auc"]], x="duration", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    sns.lineplot(data=LRSSS[["seed", "duration", "roc_auc"]], x="duration", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    sns.lineplot(data=HORPCA[["seed", "duration", "roc_auc"]], x="duration", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    plt.ylim(0.7,0.95 )
    plt.title("AUC for varying duration with NoG = 450, r = 2")
    plt.xlabel("Duration")
    plt.ylabel("AUC")
    #plt.tight_layout()
    plt.grid(True)
   

    plt.subplot(1, 2, 2) 

    
    sns.lineplot(data=LRSTSS[["seed", "duration", "F1_score"]], x="duration", y="F1_score", label='LRSTSS',marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSTS[["seed", "duration", "F1_score"]], x="duration", y="F1_score", label='LRSTS',marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSSS[["seed", "duration", "F1_score"]], x="duration", y="F1_score", label='LRSSS',marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=HORPCA[["seed", "duration", "F1_score"]], x="duration", y="F1_score", label='HORPCA',marker = 'o',errorbar= ('ci', 95))
    plt.ylim(0.2,0.85 )
    plt.title("F1_score for varying duration with NoG = 450, r = 2")
    plt.xlabel("Duration")
    plt.ylabel("F1_score")
    #plt.tight_layout()
    plt.grid(True)
    plt.legend(loc = 'best', bbox_to_anchor=(1, 1))
    
    
    
    
    plt.show()
```
