```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
LRSTSS = pd.read_csv("LRSTSS_varying_rad.csv")
LRSTS = pd.read_csv("LRSTS_varying_rad.csv")
HORPCA = pd.read_csv("HORPCA_varying_rad.csv")
LRSSS =  pd.read_csv("LRSSS_varying_rad.csv")
```

```python
    
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1) 
    sns.lineplot(data=LRSTSS[["seed", "radius", "roc_auc"]], x="radius", y="roc_auc",marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSTS[["seed", "radius", "roc_auc"]], x="radius", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    sns.lineplot(data=LRSSS[["seed", "radius", "roc_auc"]], x="radius", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    sns.lineplot(data=HORPCA[["seed", "radius", "roc_auc"]], x="radius", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    plt.ylim(0.7,0.95 )
    plt.title("AUC for varying radius with NoG = 100, d = 4")
    plt.xlabel("Radius")
    plt.ylabel("AUC")
    #plt.tight_layout()
    plt.grid(True)
   

    plt.subplot(1, 2, 2) 

    
    sns.lineplot(data=LRSTSS[["seed", "radius", "F1_score"]], x="radius", y="F1_score",marker = 'o', label='LRSTSS',errorbar= ('ci', 95))
    sns.lineplot(data=LRSTS[["seed", "radius", "F1_score"]], x="radius", y="F1_score", marker = 'o',label='LRSTS',errorbar= ('ci', 95))
    sns.lineplot(data=LRSSS[["seed", "radius", "F1_score"]], x="radius", y="F1_score", marker = 'o',label='LRSSS',errorbar= ('ci', 95))
    sns.lineplot(data=HORPCA[["seed", "radius", "F1_score"]], x="radius", y="F1_score", marker = 'o',label='HORPCA',errorbar= ('ci', 95))
    plt.ylim(0.2,0.85 )
    plt.title("F1_score for varying radius with NoG = 100, d = 4")
    plt.xlabel("Radius")
    plt.ylabel("F1_score")
    #plt.tight_layout()
    plt.grid(True)
    plt.legend(loc = 'best', bbox_to_anchor=(1, 1))
    
    
    
    
    plt.show()
```
