```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
LRSTSS = pd.read_csv("LRSTSS_varying_snr.csv")
LRSTS = pd.read_csv("LRSTS_varying_snr.csv")
HORPCA = pd.read_csv("HORPCA_varying_snr.csv")
LRSSS =  pd.read_csv("LRSSS_varying_snr.csv")
```

```python
    
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1) 
    sns.lineplot(data=LRSTSS[["seed", "snr", "roc_auc"]], x="snr", y="roc_auc",marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSTS[["seed", "snr", "roc_auc"]], x="snr", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    sns.lineplot(data=LRSSS[["seed", "snr", "roc_auc"]], x="snr", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    sns.lineplot(data=HORPCA[["seed", "snr", "roc_auc"]], x="snr", y="roc_auc",marker = 'o',errorbar = ('ci', 95))
    plt.title("AUC for varying SNR with r = 2, d = 10, noa = 450")
    plt.xlabel("SNR")
    plt.ylabel("AUC")
    #plt.tight_layout()
    plt.grid(True)
   

    plt.subplot(1, 2, 2) 

    
    sns.lineplot(data=LRSTSS[["seed", "snr", "F1_score"]], x="snr", y="F1_score", label='LRSTSS',marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSTS[["seed", "snr", "F1_score"]], x="snr", y="F1_score", label='LRSTS',marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=LRSSS[["seed", "snr", "F1_score"]], x="snr", y="F1_score", label='LRSSS',marker = 'o',errorbar= ('ci', 95))
    sns.lineplot(data=HORPCA[["seed", "snr", "F1_score"]], x="snr", y="F1_score", label='HORPCA',marker = 'o',errorbar= ('ci', 95))
    plt.title("F1_score for varying groups with r = 2, d = 10, noa = 450")
    plt.xlabel("SNR")
    plt.ylabel("F1_score")
    #plt.tight_layout()
    plt.grid(True)
    plt.legend(loc = 'best', bbox_to_anchor=(1, 1))
    
    
    
    
    plt.show()
```
